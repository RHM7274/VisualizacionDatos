```python
df = pd.read_excel('datasets/Canada.xlsx',
                   sheet_name='Canada by Citizenship',
                   engine = 'openpyxl',
                   skiprows = range(20),
                   skipfooter=2 #saltarse los últimos renglones
                  )
df
```

```python
df.columns
```

```python
all(isinstance(columns,str)for columns in df.columns) #No todas las columnas en el df son string
```

```python
df.columns = list(map(str,df.columns)) #Para cambiar todas las columnas a str
all(isinstance(columns,str)for columns in df.columns) #Para comprobar que todas las columnas son str
```

```python
df.drop(['Type', 'Coverage','REG', 'DEV', 'DevName','Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46',
       'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],inplace=True, axis= 1)
df.columns
```

```python
df.rename(columns={'OdName':'Country',
                  
                  'AreaName':'Continent',
                  'RegName':'Region',
                  },
          inplace= True)
df.columns #cambiar el nombre de las columnas
```

```python
df['Total']=df.iloc[:,4:].sum(axis=1)
df.head()
```

```python
df.set_index('Country',inplace=True)
df.head
```

```python
df.loc['Japan','Total']
```

```python
years =list(map(str,range(1980,2014)))
years #se creó una nueva variable llamada years desde 1980 a 2014
```

```python
df.loc['Japan',years].sum() #se muestran los valores de JAPAN de la nueva variable creada
```

```python
df.loc['Japan',years].plot()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('ggplot')
```

```python
df.loc['Japan',years].plot() #hacer un tipo de gráfica
```

# Me gustaría saber cuantas personas llegaroo desde Haiti a Canada desde 1980-2013 con una gráfica de lineas

```python
df.loc['Haiti',years].plot()
plt.xlabel('X',fontsize=16)
plt.xticks(fontsize=16,rotation=45)

plt.ylabel('Número de inmigrantes', fontsize=16)
plt.yticks(fontsize=16, rotation=45)

plt.title('Número de inmigrantes desde Haiti a Canada desde 1980-2013',fontsize=13)
plt.text(10,6300,'Terremoto')
plt.show()
```
