```python
df=pd.read_csv('datasets/census.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot=pd.pivot_table(df,
                        index='STNAME'
                       )
my_pivot #variable para hacer la tabla pivote de la variable 'STNAME'
```

```python
my_pivot=pd.pivot_table(df,
                        index='STNAME',
                        values='BIRTHS2010'
                       )
my_pivot #variable para hacer la tabla pivote de la variable 'STNAME' y 'BIRTHS2010'
#automaticamente te saca los promedios de la variable pero puede sacar más funciones con "aggfunc='median'"
```

```python
my_pivot=pd.pivot_table(df,
                        index='STNAME',
                        values='BIRTHS2010',
                        aggfunc=['std','sum','median','mean']
                       )
my_pivot
```

```python
my_pivot=pd.pivot_table(df,
                        index=['STNAME','CTYNAME'],
                        values='BIRTHS2010',
                        #aggfunc=['std','sum','median','mean']
                       )
my_pivot
```

```python
my_pivot=pd.pivot_table(df,
                        index=['STNAME','CTYNAME'],
                        values=['BIRTHS2010','BIRTHS2011'],
                        aggfunc='sum',
                        columns='SUMLEV',
                        margins=True
                       )
my_pivot
#para poner columnas se usa 'columns'
#si quieres subíndices de  las filas poner index=['STNAME','CTYNAME']
#margins=True  sirve para poner una columna  con la sumatoria de todo
```
