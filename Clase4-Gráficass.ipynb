```python
#Para generar múltiples columnas para separar los datos
fig = plt.figure()
ax1 = fig.add_subplot(1 , 2 , 1) #1 column, 2 row, position 1
sns.boxplot(data = df.loc[df['Gender'] == 'M', ['Score']]).set_title('Box Plot de género M')
ax2 = fig.add_subplot(1,2,2) #1 column, 2 row, position 2
sns.boxplot(data = df.loc[df['Gender'] == 'F', ['Score']]).set_title('Box Plot de género F')
```
