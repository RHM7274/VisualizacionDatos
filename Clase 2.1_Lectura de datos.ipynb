### loc y iloc
[row_inicial:row_final,col_inicial:col_final]
si usa loc, usa nombres de las renglones y nombres de las columnas

si usa iloc, usa indices de las renglones y las columnas

```python
df.iloc[2:10, 3:6] #el primero es los números de filas que vas a llamar, el segundo es  el nombre de las columnas
```

```python
df.loc[2:7,'Age':'Height']
```
